```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.stats import norm

plt.rcParams['font.size'] = '16'
```

```python
u = np.arange(start=0, stop=2, step=0.05)
A = 3

def u2a(u=u, A=A):
    return (np.exp(-A*u) - 1) / (np.exp(-A) - 1)

a = u2a(u)

fig, ax = plt.subplots()
fig.patch.set_facecolor('white')
ax.plot(u, a)
ax.set_xlabel(r'$u$')
ax.set_ylabel(r'$a$')
ax.set_title('Nonlinear activation')
```
